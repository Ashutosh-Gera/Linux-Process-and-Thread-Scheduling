```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df1 = pd.read_csv("thrA.txt", sep=" ", header=None)
df2 = pd.read_csv("thrB.txt", sep=" ", header= None)
df3 = pd.read_csv("thrC.txt", sep=" ", header=None)

df1 = df1.drop([1], axis=1)
df2 = df2.drop([1], axis=1)
df3 = df3.drop([1], axis=1)

fig = plt.figure(figsize = (25, 10))
plt.bar(df1[2], df1[0], width=1.9)
plt.bar(df2[2], df2[0], width=1.9)
plt.bar(df3[2], df3[0], width=1.9)
plt.legend(["Thread A", "Thread B","Thread C"])
plt.xlabel("Priority Values", fontsize=25)
plt.ylabel("Time", fontsize=25)
plt.show()
```
